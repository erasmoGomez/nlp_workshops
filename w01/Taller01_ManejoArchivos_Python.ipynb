### üìÑ Ejemplo: Procesamiento del texto extra√≠do del PDF

```python
import re
```

```python
# Tokenizar oraciones
oraciones = re.split(r'[.!?]\s+', texto_pdf)
print("Primera oraci√≥n:", oraciones[0])

# Tokenizar palabras
palabras = re.findall(r'\b\w+\b', texto_pdf.lower())
print("Primeras 10 palabras:", palabras[:10])
```

### ‚úÖ Generar audio en espa√±ol con gTTS en Google Colab

```python
!pip install gTTS pydub

from gtts import gTTS
from pydub import AudioSegment
```

```python
texto = "Este es un ejemplo de audio generado autom√°ticamente para an√°lisis."

# Crear archivo MP3
tts = gTTS(text=texto, lang='es')
tts.save("audio_espanol.mp3")
```

### ‚úÖ Generar video con ese audio:





```python
from moviepy.editor import ColorClip, AudioFileClip
```

```python


# Crear el audio con gTTS
texto = "Este es un ejemplo de audio generado autom√°ticamente para an√°lisis."
tts = gTTS(text=texto, lang='es')
tts.save("audio_espanol.mp3")

# Obtener duraci√≥n del audio
audio = AudioSegment.from_mp3("audio_espanol.mp3")
audio.export("audio_espanol.wav", format="wav")
duracion = audio.duration_seconds

# Crear un clip de video con fondo blanco
video_clip = ColorClip(size=(640, 480), color=(255, 255, 255), duration=duracion)

# Cargar audio y unir al clip
audio_clip = AudioFileClip("audio_espanol.wav")
video_clip = video_clip.set_audio(audio_clip)

# Exportar el video
video_clip.write_videofile("video_espanol_wav.mp4", fps = 60, codec="libx264", audio_codec="aac")
```

### üîä Cargar archivos de audio (`.wav`, `.mp3`) para an√°lisis de habla

```python
!pip install librosa

import librosa
```

```python

audio, sr = librosa.load("audio_espanol.wav", sr=None)

print(f"Duraci√≥n: {len(audio)/sr:.2f} segundos - Sample Rate: {sr}")
```

### üéß Visualizar el espectrograma

```python
import librosa.display
import matplotlib.pyplot as plt
```

```python

#CARGA DEL ARCHIVO
audio, sr = librosa.load("audio_espanol.wav", sr=None)

#CONFIGURACION DEL GRAFICO
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio, sr=sr) #MUESTRA INFO DEL ESPECTOGRAMA
plt.title("Forma de onda del audio")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.tight_layout()
#DISPLAY
plt.show()
```

### üîä Ejemplo: Extraer caracter√≠sticas ac√∫sticas (MFCC) del audio

```python
mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)

import matplotlib.pyplot as plt
import librosa.display

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title("MFCC del audio")
plt.tight_layout()
plt.show()
```

## Una opcion es usar OCR

```python
!apt-get install tesseract-ocr -y
!apt-get install tesseract-ocr-spa -y
!pip install pytesseract Pillow

import pytesseract
from PIL import Image
import fitz  # PyMuPDF
import io
```

```python
# Abrir el PDF y extraer imagen de la p√°gina 10
doc = fitz.open(ruta + '/Amahuaca_dictionary_Setiembre_2023.pdf')
page = doc.load_page(10)
pix = page.get_pixmap(dpi=300)
img = Image.open(io.BytesIO(pix.tobytes("png")))
```

```python
img
```

```python
# Realizar OCR
texto = pytesseract.image_to_string(img, lang="spa")
```
